```python
#считываем тренировочные данные
train = pd.read_csv('train.csv')
formula_train = pd.read_csv('formula_train.csv')

print(f'train dataset shape: {train.shape}')
print(f'formula_train dataset shape: {formula_train.shape}')
```

```python
#считываем тестовые данные
test = pd.read_csv('test.csv')
formula_test = pd.read_csv('formula_test.csv')

print(f'test dataset shape: {test.shape}')
print(f'formula_test dataset shape: {formula_test.shape}')
```

```python
#объединяю train и formula_train для удобства дальнейшей работы
formula_train = formula_train.drop(columns=['critical_temp'])
train_full = pd.concat([train, formula_train], axis=1)
print(train_full.shape)
```

```python
#из анализа данных понятно, что для дальнейше работы нам не понадобится признак "material"
train_full.drop(columns = ['material'], inplace=True)
```

```python
train_full.head()
```

```python
#определим вектор признаков и вектор ответов
X = train_full.drop(columns=['critical_temp'])
y = train_full['critical_temp']

print(X.shape)
print(y.shape)
```

```python
#производим теже самые действия по обработке данных над тестовым набором данных
test_full = pd.concat([test, formula_test], axis=1)
print(test_full.shape)
```

```python
test_full.drop(columns = ['material'], inplace=True)
print(test_full.shape)
```

```python
#разделяю на тренировочную и тестовую подвыборки датасет train
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#обучаю модель, использую метрики: средняя абсолютная ошибка, средняя квадратичная ошибка и коэффициент детерминации
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#создадим модель линейной регрессии
model = LinearRegression()

#обучим модель на тренировочном наборе данных
model.fit(X_train, y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
featurs = test_full.columns
```

```python
coeff_df = pd.DataFrame(model.coef_, columns=['Coeff'])
coeff_df['featurs'] = featurs
```

```python
coeff_df.sort_values(by='Coeff')
```

```python
#предскажем результат для тех же самых тренировочных данных
y_pred = model.predict(X_test)

print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
